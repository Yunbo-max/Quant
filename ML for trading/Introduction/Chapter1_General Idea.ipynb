# Information Coefficient (IC) and Information Ratio (IR)

## Information Coefficient (IC):

The Information Coefficient (IC) is a statistical measure that assesses the quality or accuracy of forecasts made in active investment management. Specifically, it quantifies the correlation between predicted returns and actual realized returns. The IC is a crucial metric because it indicates how well the forecasts align with the subsequent market outcomes.

In mathematical terms, the Information Coefficient (IC) is represented by the correlation coefficient (œÅ) between the predicted returns (P) and the actual returns (A) over a specified time period:

\[ \text{IC} = \text{Corr}(P, A) \]

A perfect positive correlation (IC = 1) signifies accurate and reliable forecasts, while an IC close to zero implies little to no predictive power in the forecasts.

## Information Ratio (IR):

The Information Ratio (IR) is a performance measure used in the field of active investment management. It quantifies the value added by active management relative to a chosen benchmark. The IR is particularly useful in assessing whether the excess returns generated by a portfolio are attributable to skillful forecasting or simply a result of increased risk.

In mathematical terms, the Information Ratio (IR) is expressed as the ratio of the expected active return (the return difference between the portfolio and the benchmark) to the volatility of this active return:

\[ \text{IR} = \frac{\text{Expected Active Return}}{\text{Volatility of Active Return}} \]

The Expected Active Return is often approximated as the product of the Information Coefficient (IC) and the square root of the number of independent bets made (N), representing the breadth of the investment strategy:

\[ \text{Expected Active Return} = \text{IC} \times \sqrt{N} \]

In summary, the Information Ratio (IR) provides a quantitative assessment of the effectiveness of active investment management by considering both the quality of forecasts (Information Coefficient) and the number of independent bets made. Higher Information Ratios generally indicate more skillful and efficient management in generating returns relative to the benchmark.
