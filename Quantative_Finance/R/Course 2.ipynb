```r
# Load required libraries
library(readr)
library(dplyr)
library(tidyr)
library(xts)

# Read the CSV file
crsp_rets <- read_csv("wrds_stock_data.csv", col_types = cols(date = col_date(format = "%Y/%m/%d")))

# Group by PERMNO and pivot wider
crsp_rets <- crsp_rets %>%
  group_by(PERMNO) %>%
  pivot_wider(names_from = PERMNO, values_from = RET)

# Convert to xts object
crsp_xts <- tk_xts(crsp_rets, select = -date, date_var = date)

# Show the resulting xts object
print(crsp_xts)
```
