```python
import yfinance as yf
from plotnine import *

# Enable inline plotting in Jupyter Notebook
%matplotlib inline

# Define a list of ten stock symbols
stock_symbols = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA',  'NFLX', 'IBM', 'INTC', 'NVDA']

# Create a subplot for each stock
for symbol in stock_symbols:
    # Download stock data
    prices = (yf.download(tickers=symbol,
                          start="2000-01-01",
                          end='2023-10-31',
                          interval='1d')
              .reset_index()
              .assign(symbol=symbol)
              .rename(columns={'Date': "Time Line",
                               'Adj Close': 'adjusted'}))
    
    print(prices.head().round(3))

    # Plot the data for the current stock
    prices_figure = (
        ggplot(prices, aes(x="Time Line")) +
        geom_line(aes(y="adjusted"), color="blue", size=1) +  # 'adjusted' line
        geom_line(aes(y="Close"), color="red", size=1) +  # 'Close' line
        labs(x="Time", y="Price", title=f"{symbol} stock prices from 2000 to 2022") +
        theme_minimal() +  # Adjust the theme to 'minimal'
        theme(figure_size=(16, 13))  # Adjust the figure size
    )

    # Print the plot for the current stock
    print(prices_figure)
```

The choice between using the close price or the adjusted price for analysis depends on the specific goals of your analysis and the factors you want to consider. Here are some guidelines on when to use each:

Close Price:

Day-to-Day Trading Analysis: If you are interested in short-term trading or intraday analysis, using the close price may be appropriate. The close price reflects the last traded price of the day and is commonly used in technical analysis, chart patterns, and short-term trading strategies.
Intraday Volatility Analysis: For intraday volatility or price movement within a single trading day, the close price is often used, as it represents the final market consensus for the day.
Example: If you are a day trader looking to make short-term trades based on intraday patterns, you may use the close price for your analysis.

Adjusted Price:

Long-Term Investment Analysis: When analyzing stocks for long-term investment, it's often advisable to use adjusted prices. Adjusted prices account for corporate actions such as dividends, stock splits, and mergers, providing a more accurate representation of a stock's true value over time.
Dividend-Adjusted Returns: If you are interested in total returns, especially when dividends are a significant part of the investment, using adjusted prices is essential. Dividend-adjusted prices reflect the impact of dividends on the stock's historical performance.
Example: If you are a long-term investor assessing the historical performance of a stock over several years, you may use adjusted prices to account for dividends and stock splits.

Special Cases:

Corporate Actions Analysis: When analyzing periods with significant corporate actions (e.g., stock splits, reverse stock splits), it may be beneficial to use adjusted prices to avoid distortions in the historical price data.
Comparing Across Securities: When comparing the performance of multiple securities or indices over an extended period, using adjusted prices helps ensure a more accurate and comparable analysis.
Example: If you are comparing the historical performance of two stocks and one of them underwent a stock split, using adjusted prices would provide a more consistent basis for comparison.

##### Question 2: Compute daily net returns for an asset of your choice and visualize the distribution of daily returns in a histogram using 100 bins. Also, use geom_vline() to add a dashed red vertical line that indicates the 5 percent quantile of the daily returns. Compute summary statistics (mean, standard deviation, minimum and maximum) for the daily returns.

```python
import yfinance as yf

prices = (yf.download(tickers = 'GOOG',
                     start = "2000-01-01",
                     end='2023-10-31',interval = '1d')
    .reset_index()
    .assign(symbol = 'AAPL')
    .rename(columns={'Date': "date",
                     'Adj Close':'adjusted'}))

returns = (prices
  .sort_values("date")
  .assign(ret=lambda x: x["adjusted"].pct_change())
  .get(["symbol", "date", "ret"])
)

returns = returns.dropna() 

print(returns.head(3))

from mizani.formatters import percent_format

quantile_05 = returns["ret"].quantile(0.05)

returns_figure = (
  ggplot(returns, aes(x="ret")) +
  geom_histogram(bins=100) +
  geom_vline(aes(xintercept=quantile_05), 
                 linetype="dashed") +
  labs(x="", y="",
       title="Distribution of daily Apple stock returns") +
  scale_x_continuous(labels=percent_format())
)


VaR = returns['ret'].quantile(0.05)
print(VaR)
#  less VaR means less Risk


print(returns['ret'].describe())

returns_figure.draw()

```

```python
# For one year using groupby rather than resampling here
(returns["ret"]
  .groupby(returns["date"].dt.year)
  .describe()
  .round(3)
)
```

```python
import numpy as np 
import pandas as pd

url = ("https://www.ssga.com/us/en/institutional/etfs/library-content/"
       "products/fund-data/etfs/us/holdings-daily-us-en-dia.xlsx")

symbols = (pd.read_excel(url, skiprows=4, nrows=30)
  .get("Ticker")
  .tolist()
)

symbols


index_prices = (yf.download(
    tickers=symbols, 
    start="2000-01-01", 
    end="2022-12-31", 
    progress=False
  )
  .melt(ignore_index=False, var_name=["variable", "symbol"])
  .reset_index()
  .pivot(index=["Date", "symbol"], columns="variable", values="value")
  .reset_index()
  .rename(columns={
    "Date": "date", 
    "Open": "open", 
    "High": "high", 
    "Low": "low",
    "Close": "close", 
    "Adj Close": "adjusted", 
    "Volume": "volume"}
  )
)

from mizani.breaks import date_breaks
from mizani.formatters import date_format
# Download historical prices for DJIA
djia_prices = yf.download(tickers='^DJI',
                          start="2000-01-01",
                          end='2022-12-31',
                          interval='1d')

# Reset index to make 'Date' a column
djia_prices = djia_prices.reset_index()

# Create a plot for DOW index constituents with DJIA line
index_prices_figure = (
    ggplot() +
    geom_line(data=index_prices, mapping=aes(y="adjusted", x="date", color="symbol")) +
    geom_line(data=djia_prices, mapping=aes(y="Adj Close", x="Date"), color="red", linetype="solid", size=0.2) +
    labs(x="", y="", color="",
         title="Stock prices of DOW index constituents with DJIA line") +
    theme(legend_position="none") +
    scale_x_datetime(date_breaks="5 years", date_labels="%Y")+
    theme_minimal() +  # Adjust the theme to 'minimal'
    theme(figure_size=(16, 13))  # Adjust the figure size
)

# Draw the plot
index_prices_figure.draw()
```

```python
import numpy as np 
import pandas as pd
import yfinance as yf
from plotnine import *

# Download historical prices for DJIA components
url = ("https://www.ssga.com/us/en/institutional/etfs/library-content/"
       "products/fund-data/etfs/us/holdings-daily-us-en-dia.xlsx")

# Read the symbols and add DJIA to the list
symbols = (pd.read_excel(url, skiprows=4, nrows=30)
  .get("Ticker")
  .tolist()
  + ['DJI']  # Add DJIA to the list
)

index_prices = (yf.download(
    tickers=symbols, 
    start="2000-01-01", 
    end="2022-12-31", 
    progress=False
  )
  .melt(ignore_index=False, var_name=["variable", "symbol"])
  .reset_index()
  .pivot(index=["Date", "symbol"], columns="variable", values="value")
  .reset_index()
  .rename(columns={
    "Date": "date", 
    "Open": "open", 
    "High": "high", 
    "Low": "low",
    "Close": "close", 
    "Adj Close": "adjusted", 
    "Volume": "volume"}
  )
)


from mizani.breaks import date_breaks
from mizani.formatters import date_format

all_returns = (index_prices
  .assign(ret=lambda x: x.groupby("symbol")["adjusted"].pct_change())
  .get(["symbol", "date", "ret"])
  .dropna(subset="ret")
)

(all_returns
  .groupby("symbol")["ret"]
  .describe()
  .round(3)
)
```

```python
# Create a plot for comparing returns
returns_plot = (
    ggplot(all_returns, aes(x="date", y="ret", color="symbol")) +
    geom_line(size=1) +
    labs(x="Date", y="Daily Returns",
         title="Comparison of Daily Returns for DJIA Components with DJIA") +
    theme_minimal() +
    theme(figure_size=(16, 8))
)

# Draw the plot
returns_plot.draw()
```
