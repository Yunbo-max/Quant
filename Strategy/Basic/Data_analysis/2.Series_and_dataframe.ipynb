# iloc and loc

```python
print(aapl_month.loc['2023-03':'2023-06', ['o', 'h', 'l', 'c']]) 
```

```python
aapl_bar = aapl_month[['o', 'h', 'l', 'c']]
above = aapl_bar[aapl_bar.c > np.mean(aapl_bar.c)]
print(above)
```

```python
aapl_bar['rate_return'] = aapl_bar.c.pct_change()
missing = aapl_bar.isnull()
print(missing)
print('---------------------------------------------')
print(missing.describe())
```

```python
print(missing[missing.rate_return == True])

drop = aapl_bar.dropna()
print(drop)
print('\n--------------------------------------------------\n')
fill = aapl_bar.fillna(0)
print(fill)
```

### Merging

```python
s1 = pd.Series([143.5, 144.09, 142.73, 144.18, 143.77], name = 'AAPL')
s2 = pd.Series([898.7, 911.71, 906.69, 918.59, 926.99], name = 'GOOG')
data_frame = pd.concat([s1, s2], axis = 1)
print(data_frame)
```

```python
log_price = np.log(aapl_bar.c)
log_price.name = 'log_price'
print(log_price)
print('\n--------------------------------------------\n')
concat = pd.concat([aapl_bar, log_price], axis = 1)
print(concat)
```
